Data Visualization

```python
# Rename columns for clarity by this , no need to repeat every time)
data.columns = ["state","date","frequency",
                "estimated_unemployment_rate","estimated_employed",
                "estimated_labour_participation_rate","region",
                "longitude","latitude","month_int"]

# Plot
plt.figure(figsize=(10,6))
sns.histplot(data=data, x="estimated_employed", hue="region", kde=False, multiple="stack")
plt.title("Estimated Employment Across Regions of India")
plt.xlabel("Estimated Employed")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=data, x="estimated_unemployment_rate", hue="region",
               multiple="stack", kde=False)
plt.title("Unemployment Rate Across Regions of India")
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Count")
plt.show()
```

```python

# Interactive scatter map of unemployment rate
fig = px.scatter_geo(
    data,
    lon="longitude",
    lat="latitude",
    color="estimated_unemployment_rate",
    hover_name="state",
    size="estimated_unemployment_rate",
    projection="natural earth",
    title="Unemployment Rate Across India (Interactive)"
)

iplot(fig)
```

```python
fig = px.scatter_matrix(
    data,
    dimensions=[
        'estimated_unemployment_rate',
        'estimated_employed',
        'estimated_labour_participation_rate'
    ],
    color='region',
    title="Scatter Matrix: Employment Metrics Across Regions of India",
    height=800,
    width=800
)

fig.update_traces(diagonal_visible=True)  # keep histograms on the diagonal
fig.show()
```

```python
import plotly.express as px

# Group by state and calculate mean unemployment rate
data_unemp = (
    data.groupby('state', as_index=False)['estimated_unemployment_rate']
    .mean()
    .sort_values('estimated_unemployment_rate', ascending=True)
)

# Bar chart
fig = px.bar(
    data_unemp,
    x='state',
    y='estimated_unemployment_rate',
    color='state',
    title='Average Unemployment Rate in Each State',
    template='plotly',
    height=600
)

# Rotate labels for clarity
fig.update_layout(
    xaxis_title="State",
    yaxis_title="Average Unemployment Rate (%)",
    xaxis_tickangle=-45,
    showlegend=False  # legend redundant since color=state
)

fig.show()
```

```python
import plotly.express as px

# Prepare data
unemployment = data[["state", "region", "estimated_unemployment_rate"]]

# Sunburst chart
fig = px.sunburst(
    unemployment,
    path=["region", "state"],
    values="estimated_unemployment_rate",
    color="estimated_unemployment_rate",
    color_continuous_scale="RdYlGn_r",  # red=high, green=low
    width=700,
    height=700,
    title="Unemployment Rate in India by Region and State"
)

fig.show()
```

```python
# Split into before and after lockdown
after_lockdown = data[(data['month_int'] >= 4) & (data['month_int'] <= 6)]
before_lockdown = data[(data['month_int'] >= 1) & (data['month_int'] <= 3)]

# Compute mean unemployment rate
af_lockdown = after_lockdown.groupby('state')['estimated_unemployment_rate'].mean().reset_index()
bef_lockdown = before_lockdown.groupby('state')['estimated_unemployment_rate'].mean().reset_index()

# Merge both
lockdown = bef_lockdown.merge(af_lockdown, on="state", suffixes=("_before", "_after"))

# Rename columns for clarity
lockdown.columns = ['state', 'unemployment_rate_before_lockdown', 'unemployment_rate_after_lockdown']

lockdown.head()
```

```python
# Calculate percentage change in unemployment after lockdown
lockdown['rate_change_in_unemployment'] = round(
    ((lockdown['unemployment_rate_after_lockdown'] - lockdown['unemployment_rate_before_lockdown'])
     / lockdown['unemployment_rate_before_lockdown']) * 100, 2
)

import plotly.express as px

# Bar chart
fig = px.bar(
    lockdown,
    x='state',
    y='rate_change_in_unemployment',
    color='rate_change_in_unemployment',
    title='Percentage Change in Unemployment Rate After Lockdown (by State)',
    template='ggplot2'
)

# Sort bars by rate change
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```
